```python
state_abbreviations = {
    "AL": "alabama",
    "AK": "alaska",
    "AZ": "arizona",
    "AR": "arkansas",
    "CA": "california",
    "CO": "colorado",
    "CT": "connecticut",
    "DE": "delaware",
    "FL": "florida",
    "GA": "georgia",
    "HI": "hawaii",
    "ID": "idaho",
    "IL": "illinois",
    "IN": "indiana",
    "IA": "iowa",
    "KS": "kansas",
    "KY": "kentucky",
    "LA": "louisiana",
    "ME": "maine",
    "MD": "maryland",
    "MA": "massachusetts",
    "MI": "michigan",
    "MN": "minnesota",
    "MS": "mississippi",
    "MO": "missouri",
    "MT": "montana",
    "NE": "nebraska",
    "NV": "nevada",
    "NH": "new hampshire",
    "NJ": "new jersey",
    "NM": "new mexico",
    "NY": "new york",
    "NC": "north carolina",
    "ND": "north dakota",
    "OH": "ohio",
    "OK": "oklahoma",
    "OR": "oregon",
    "PA": "pennsylvania",
    "RI": "rhode island",
    "SC": "south carolina",
    "SD": "south dakota",
    "TN": "tennessee",
    "TX": "texas",
    "UT": "utah",
    "VT": "vermont",
    "VA": "virginia",
    "WA": "washington",
    "WV": "west virginia",
    "WI": "wisconsin",
    "WY": "wyoming",

    # doofus decides that DC is a state.
    "DC": "district of columbia",
}
caps_state_name = {
    "AL": "Alabama",
    "AK": "Alaska",
    "AZ": "Arizona",
    "AR": "Arkansas",
    "CA": "California",
    "CO": "Colorado",
    "CT": "Connecticut",
    "DE": "Delaware",
    "FL": "Florida",
    "GA": "Georgia",
    "HI": "Hawaii",
    "ID": "Idaho",
    "IL": "Illinois",
    "IN": "Indiana",
    "IA": "Iowa",
    "KS": "Kansas",
    "KY": "Kentucky",
    "LA": "Louisiana",
    "ME": "Maine",
    "MD": "Maryland",
    "MA": "Massachusetts",
    "MI": "Michigan",
    "MN": "Minnesota",
    "MS": "Mississippi",
    "MO": "Missouri",
    "MT": "Montana",
    "NE": "Nebraska",
    "NV": "Nevada",
    "NH": "New Hampshire",
    "NJ": "New Jersey",
    "NM": "New Mexico",
    "NY": "New York",
    "NC": "North Carolina",
    "ND": "North Dakota",
    "OH": "Ohio",
    "OK": "Oklahoma",
    "OR": "Oregon",
    "PA": "Pennsylvania",
    "RI": "Rhode Island",
    "SC": "South Carolina",
    "SD": "South Dakota",
    "TN": "Tennessee",
    "TX": "Texas",
    "UT": "Utah",
    "VT": "Vermont",
    "VA": "Virginia",
    "WA": "Washington",
    "WV": "West Virginia",
    "WI": "Wisconsin",
    "WY": "Wyoming",

    # doofus decides that DC is a state.
    "DC": "District of Columbia",
}
leaseData = pd.read_csv("./Leases.csv", encoding="latin1").drop_duplicates()
majorMarketData = pd.read_csv("./Major Market Occupancy Data-revised.csv", encoding="latin1")
priceData = pd.read_csv("./Price and Availability Data.csv", encoding="latin1")
unemploymentData = pd.read_csv("./Unemployment.csv", header=0, encoding="latin1")
populationData = pd.read_csv("./population.csv", encoding="latin1")

populationData["Resident Population Density"] = populationData["Resident Population Density"].str.replace(",", "").astype(float)

maxStatePopulation = np.nanmax(populationData["Resident Population Density"])

#print(populationData)

def get_general_category(industry: str):
    for current_broad_category_dict in broad_categories:
        if industry in current_broad_category_dict.get("subcategories"):
            return current_broad_category_dict.get("category")

def get_competition_quotient(state_abbr, industry):

    # american companies only support full name. conversion needed here
    state_name = state_abbreviations.get(state_abbr)
    general_category = get_general_category(industry)

    companies_in_state = americanCompanies[americanCompanies["region"] == state_name]
    companies_in_general_industry_in_state = companies_in_state[
        companies_in_state["industry"].apply(lambda x: get_general_category(x) == general_category)
    ]

    num_companies_in_state = len(companies_in_state)
    num_companies_in_industry_category_in_state = len(companies_in_general_industry_in_state)

    #print("Num of companies in industry category in state: ", num_companies_in_industry_category_in_state)
    #print("num companies in state: ", num_companies_in_state)
    if (num_companies_in_state == 0 or num_companies_in_industry_category_in_state == 0):
        return 0
    percentage = num_companies_in_industry_category_in_state / num_companies_in_state


    return lerp(0, 0.5, percentage)

def lerp(x1,y1,t):
  return x1 + t * (y1 - x1)


def get_price_quotient(state_abbr):

    state_data = leaseData[leaseData["state"] == state_abbr]

    stateRent = state_data["overall_rent"];
    nationRent = leaseData["overall_rent"];

    if(stateRent.size <= 0 or nationRent.size <= 0):
        return 0

    mean_rent_nation = np.nanmean(nationRent)
    mean_rent = np.nanmean(stateRent)


    return lerp(0 , 0.5, mean_rent/mean_rent_nation)

def get_space_quotient(state_abbr):

    state_data = leaseData[leaseData["state"] == state_abbr]

    avgNationSpace = np.nanmean(leaseData["available_space"].to_numpy())

    maxStateSpace = np.nanmax(state_data["available_space"].to_numpy())

    return (avgNationSpace / maxStateSpace)

def get_population_quotient(state_abbr):

    if state_abbr not in caps_state_name:
        print("no state dog")
        return 0

    statePopData = populationData[populationData["Name"] == caps_state_name[state_abbr]];
    stateCurrent = statePopData[statePopData["Year"] == 2020]

    stateDensity = stateCurrent.iloc[0]["Resident Population Density"]

    return (float(stateDensity)/maxStatePopulation)*10


def get_favorability(pref,quotient):
    return (1-abs((pref-quotient))) * 100

def state_favorability_score(state_abbr, industry, competition_preference, price_preference, population_preference, space_preference):
    competition_quotient: float = get_competition_quotient(state_abbr, industry)
    price_quotient: float = get_price_quotient(state_abbr)
    population_quotient: float = get_population_quotient(state_abbr)
    space_quotient: float = get_space_quotient(state_abbr)

    print("Quotients for state " + state_abbr + ": " + "\n\tCompetition:"+str(competition_quotient)+ "\n\tPrice:"+str(price_quotient)+"\n\tpopulation:"+str(population_quotient), "\n\tSpace:"+str(space_quotient));

    comp_score = get_favorability(competition_preference, competition_quotient)
    price_score = get_favorability(price_preference, price_quotient)
    population_score = get_favorability(population_preference, population_quotient)
    space_score = get_favorability(space_preference, space_quotient)

   #print("\nScores for state " + state_abbr + ": " + "\n\tCompetition:"+str(comp_score)+ "\n\tPrice:"+str(price_score)+"\n\tpopulation:"+
   #       str(population_score), "\n\tSpace:"+str(space_score));



    return np.mean([comp_score, price_score, population_score, space_score])

print("Favorability TX: ", str(state_favorability_score("TX", "entertainment", 0.6, 0.7, 0.9, 0.6)))

print("Favorability NY: ", str(state_favorability_score("NY", "entertainment", 0.6, 0.7, 0.9, 0.6)))
```

```python
for state in states:
    # Initialize a dictionary to store preferability scores for each industry
    industry_scores = {}

    if state == "CO":
        continue

    # Technology & Information Systems: High competition, moderate price sensitivity, urban population preference, low space needs.
    tech_score = state_favorability_score(state, "Technology & Information Systems", 0.8, 0.5, 0.9, 0.2)
    industry_scores["Technology & Information Systems"] = tech_score

    # Healthcare & Life Sciences: Moderate competition, low price sensitivity, population-dependent, moderate space needs.
    health_score = state_favorability_score(state, "Healthcare & Life Sciences", 0.6, 0.3, 0.7, 0.5)
    industry_scores["Healthcare & Life Sciences"] = health_score

    # Manufacturing & Industrial Production: High competition, high price sensitivity, suburban/rural preference, high space needs.
    manuf_score = state_favorability_score(state, "Manufacturing & Industrial Production", 0.7, 0.9, 0.4, 0.8)
    industry_scores["Manufacturing & Industrial Production"] = manuf_score

    # Finance, Business & Professional Services: High competition, moderate price sensitivity, urban preference, low space needs.
    finance_score = state_favorability_score(state, "Finance, Business & Professional Services", 0.9, 0.6, 0.9, 0.3)
    industry_scores["Finance, Business & Professional Services"] = finance_score

    # Government, Public Sector & Social Services: Low competition, low price sensitivity, population-dependent, moderate space needs.
    gov_score = state_favorability_score(state, "Government, Public Sector & Social Services", 0.2, 0.2, 0.8, 0.5)
    industry_scores["Government, Public Sector & Social Services"] = gov_score

    # Education & Research: Moderate competition, low price sensitivity, urban/suburban preference, moderate space needs.
    edu_score = state_favorability_score(state, "Education & Research", 0.5, 0.3, 0.7, 0.6)
    industry_scores["Education & Research"] = edu_score

    # Consumer Goods, Retail & Hospitality: High competition, high price sensitivity, high population preference, moderate space needs.
    retail_score = state_favorability_score(state, "Consumer Goods, Retail & Hospitality", 0.8, 0.8, 1.0, 0.5)
    industry_scores["Consumer Goods, Retail & Hospitality"] = retail_score

    # Media, Arts & Entertainment: High competition, moderate price sensitivity, urban preference, low space needs.
    media_score = state_favorability_score(state, "Media, Arts & Entertainment", 0.7, 0.5, 0.9, 0.3)
    industry_scores["Media, Arts & Entertainment"] = media_score

    # Transportation, Logistics & Supply Chain: Moderate competition, high price sensitivity, strategic location preference, high space needs.
    transport_score = state_favorability_score(state, "Transportation, Logistics & Supply Chain", 0.6, 0.8, 0.5, 0.9)
    industry_scores["Transportation, Logistics & Supply Chain"] = transport_score

    # Energy, Environment & Utilities: Low competition, high price sensitivity, rural/resource-rich preference, high space needs.
    energy_score = state_favorability_score(state, "Energy, Environment & Utilities", 0.3, 0.9, 0.3, 0.8)
    industry_scores["Energy, Environment & Utilities"] = energy_score

    # Agriculture, Food & Natural Resources: Moderate competition, high price sensitivity, rural preference, very high space needs.
    agri_score = state_favorability_score(state, "Agriculture, Food & Natural Resources", 0.5, 0.9, 0.2, 1.0)
    industry_scores["Agriculture, Food & Natural Resources"] = agri_score

    # Construction, Engineering & Infrastructure: Moderate competition, high price sensitivity, population/growth-dependent, moderate space needs.
    constr_score = state_favorability_score(state, "Construction, Engineering & Infrastructure", 0.6, 0.8, 0.6, 0.7)
    industry_scores["Construction, Engineering & Infrastructure"] = constr_score

    # Print the results for the current state
    print(f"\nPreferability Scores for {state}:")
    for industry, score in industry_scores.items():
        print(f"{industry}: {score:.2f}")
```
